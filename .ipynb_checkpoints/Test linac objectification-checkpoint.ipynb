```python
from lxml import objectify
```

```python
xml = open('linacLegoParsed.xml', 'r')
root = objectify.parse(xml)
xml.close()
```

```python
linac = root.find('linac')
linacDataIds = [ data.get('id') for data in linac.linacData.d ]
sectionIds = [ section.get('id') for section in linac.section ]
cellIds = [ cell.get('id') for cell in linac.section.cell ]
```

```python
#section = linac.section[1]
for sectionId in sectionIds:
    print sectionId
    print linac.xpath("//section[@id='"+sectionId+"']")
```

```python
def getSection(sectionId):
    section = linac.xpath("//section[@id='"+sectionId+"']")[0]
    return section
```

```python
section = getSection('HEBT')
section.get('rfHarmonic')
```

```python
def getCell(sectionId,cellId):
    section = getSection(sectionId)
    cell = section.xpath("//cell[@id='"+cellId+"']")[0]
    return cell
```

```python
cell = getCell('HEBT','030')
#for data in cell.d:
#    print data.get('id')
#    print data
cell.xpath("//d[@id='g1']")
```

```python
def getCellDatum(sectionId,cellId,datumId):
    section = getSection(sectionId)
    datum = section.xpath("//cell[@id='"+cellId+"']/d[@id='"+datumId+"']")[0]
    cell = getCell(sectionId,cellId)
    datum = cell.xpath("//d[@id='"+datumId+"']")[0]
    return datum
```

```python
!rm example1_graph.png

import pydot

# Create a new graph with pydot.Dot()
graph = pydot.Dot(graph_type='graph')

# let's add the relationship between the linac and sections
for sectionId in sectionIds:
    # node, they are just strings like you can see
    edge = pydot.Edge("linac", sectionId)
    # Add the edge to our graph
    graph.add_edge(edge)

for sectionId in sectionIds:
    for cellId in cellIds:
        edge = pydot.Edge(sectionId, cellId)
        graph.add_edge(edge)

# Save the graph into a file
graph.write_png('example1_graph.png')

# Disply the image in the notebook
from IPython.display import Image
Image('example1_graph.png')
```

```python
from numpy import sin, pi, linspace
from nvd3 import lineChart, scatterChart
import random
 
type = "scatterChart"
# I modified the example at the end of the following line
chart = scatterChart(name=type, height=350, x_is_date=False, assets_directory="rawgithub")
chart.set_containerheader("\n\n<h2>" + type + "</h2>\n\n")
nb_element = 50
xdata = [i + random.randint(1, 10) for i in range(nb_element)]
ydata = [i * random.randint(1, 10) for i in range(nb_element)]
ydata2 = [x * 2 for x in ydata]
ydata3 = [x * 5 for x in ydata]
 
kwargs1 = {'shape': 'circle', 'size': '1'}
kwargs2 = {'shape': 'cross', 'size': '10'}
kwargs3 = {'shape': 'triangle-up', 'size': '100'}
 
extra_serie = {"tooltip": {"y_start": "", "y_end": " calls"}}
chart.add_serie(name="serie 1", y=ydata, x=xdata, extra=extra_serie, **kwargs1)
chart.add_serie(name="serie 2", y=ydata2, x=xdata, extra=extra_serie, **kwargs2)
chart.add_serie(name="serie 3", y=ydata3, x=xdata, extra=extra_serie, **kwargs3)
 
chart.buildhtml()
output_file = open('test_scatterChart.html', 'w')
output_file.write(chart.htmlcontent)
output_file.close()

from IPython.core.display import HTML
HTML("test_scatterChart.html")
```
