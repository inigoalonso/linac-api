```python
import urllib, json
```

```python
baseUrl = 'http://127.0.0.1:5000/api/v1/'
```

```python
linacUrl = baseUrl+'linac'
response = urllib.urlopen(linacUrl);
data = json.loads(response.read())
data
```

```python
sectionsUrl = baseUrl+'linac/getSections'
response = urllib.urlopen(sectionsUrl);
sectionsData = json.loads(response.read())
sections = sectionsData['Sections']
```

```python
for section in sections:
    print(section)
    sectionUrl = baseUrl+'linac/'+section+'/getCells'
    response = urllib.urlopen(sectionUrl);
    cellsData = json.loads(response.read())
    #print(cellsData)
    cells = cellsData['Cells']
    for cell in cells:
        print('    '+cell)
```
